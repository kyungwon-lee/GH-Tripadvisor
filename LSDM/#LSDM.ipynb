```python
import re

import nltk
nltk.download('wordnet')
from nltk.corpus import stopwords
nltk.download('stopwords')
stopwords.words('english')
stopword_list=stopwords.words('english')
stopword_list.append('hotel')

from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

print(len(review_body))

finallist= []
for i in range(0, len(review_body)):
    

    # 1. html 제거
    review = re.sub('<.+?>', '', str(review_body[i]), 0)  
    #print(review_body)

    # 2. 이외의 숫자, 특수문자 제거
    review = re.sub('[^a-zA-Z]', ' ', review)
    #print(review_body)

    # 3. 소문자화
    review = review.lower()
    #print(review_body)

    # 4. tokenize
    review  = review.split()
    #print(review_body)

    # 5. stopword, 한글자 제거
    

    final =[]
    for w in review:
        if w not in stopword_list:
            if len(w)>1:
                final.append(w)
            else : pass
        else: pass    
    #print(final)


    # 6. lemmatizing
    final_list = []
    for o in final:
        z = lemmatizer.lemmatize(o)
        final_list.append(z)
    #print(final_list)

    finallist.append(final_list)

all_j = []
each_j = []

for i in range(len(finallist)):
    #print(finallist[i])
    all_j += finallist[i]
    each_j.append(finallist[i])
    
print(all_j)    
```

```python
selected_words = [f[0] for f in text.vocab().most_common(1000)]

#print(selected_words)

def term_frequency(doc):
    return [doc.count(word) for word in selected_words]


print(term_frequency(each_j[1]))

#train_x = [term_frequency(d) for d, _ in train_docs]
#test_x = [term_frequency(d) for d, _ in test_docs]
#train_y = [c for _, c in train_docs]
#test_y = [c for _, c in test_docs]

```

```python
train_x = []
train_y = []
test_x = []
test_y = []

for i in range(800):
    train_x.append(term_frequency(each_j[i]))
    train_y.append([1, 0, 0, 0, 0])
for i in range(800):
    train_x.append(term_frequency(each_j[i+1000]))
    train_y.append([0, 1, 0, 0, 0])
for i in range(800):
    train_x.append(term_frequency(each_j[i+2000]))
    train_y.append([0, 0, 1, 0, 0])
for i in range(800):
    train_x.append(term_frequency(each_j[i+3000]))
    train_y.append([0, 0, 0, 1, 0])
for i in range(800):
    train_x.append(term_frequency(each_j[i+4000]))
    train_y.append([0, 0, 0, 0, 1])
    
for i in range(200):    
    test_x.append(term_frequency(each_j[i+800]))
    test_y.append([1, 0, 0, 0, 0])
for i in range(200):    
    test_x.append(term_frequency(each_j[i+1800]))
    test_y.append([0, 1, 0, 0, 0])
for i in range(200):    
    test_x.append(term_frequency(each_j[i+2800]))
    test_y.append([0, 0, 1, 0, 0])
for i in range(200):    
    test_x.append(term_frequency(each_j[i+3800]))
    test_y.append([0, 0, 0, 1, 0])
for i in range(200):    
    test_x.append(term_frequency(each_j[i+4800]))    
    test_y.append([0, 0, 0, 0, 1])
    
print(train_x)
print(train_y)
print(test_x)
print(test_y)
    
```

```python
import numpy as np

x_train = np.asarray(train_x).astype('float32')
x_test = np.asarray(test_x).astype('float32')

y_train = np.asarray(train_y).astype('float32')
y_test = np.asarray(test_y).astype('float32')
```

```python
from tensorflow.keras import models
from tensorflow.keras import layers
from tensorflow.keras import optimizers
from tensorflow.keras import losses
from tensorflow.keras import metrics

model = models.Sequential()
model.add(layers.Dense(64, activation='relu', input_shape=(1000,)))
model.add(layers.Dense(64, activation='relu'))
#model.add(layers.Dense(5, activation='sigmoid'))
model.add(layers.Dense(5, activation='softmax'))

#model.compile(optimizer=optimizers.RMSprop(lr=0.001), loss=losses.binary_crossentropy, metrics=[metrics.binary_accuracy])
model.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])

model.fit(x_train, y_train, epochs=20, batch_size=512)
results = model.evaluate(x_test, y_test)
```

```python
import tensorflow as tf
import numpy as np

X = tf.placeholder(tf.float32, [None, 1000])
Y = tf.placeholder(tf.float32, [None, 5])
keep_prob = tf.placeholder(tf.float32)

W1 = tf.Variable(tf.random_normal([1000, 256]))
L1 = tf.nn.relu(tf.matmul(X, W1))
L1 = tf.nn.dropout(L1, keep_prob)

W2 = tf.Variable(tf.random_normal([256, 256]))
L2 = tf.nn.relu(tf.matmul(L1, W2))
L2 = tf.nn.dropout(L2, keep_prob)

W3 = tf.Variable(tf.random_normal([256, 5]))
model = tf.matmul(L2, W3)

cost = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(logits=model, labels=Y))
optimizer = tf.train.AdamOptimizer(0.001).minimize(cost)

init = tf.global_variables_initializer()
sess = tf.Session()
sess.run(init)

#batch_size = 100
#total_batch = int()

#total_cost = 0
for epoch in range(1000):
    
    _, costval = sess.run([optimizer, cost], feed_dict={X:x_train, Y:y_train, keep_prob:0.8})
    #total_cost += cost_val
    print(epoch+1, costval)
    
    if epoch % 100 == 0:
        is_correct = tf.equal(tf.argmax(model, 1), tf.argmax(Y, 1))
        accuracy = tf.reduce_mean(tf.cast(is_correct, tf.float32))
        print(sess.run(accuracy, feed_dict={X:x_test, Y:y_test, keep_prob:1}))    

    
```
