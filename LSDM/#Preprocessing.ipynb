```python
import re

import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('wordnet')
stopwords.words('english')
stopword_list=stopwords.words('english')
stopword_list.append('hotel')

from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()


finallist= []
for i in range(0, len(review_body)):
    

    # 1. html 제거
    review = re.sub('<.+?>', '', str(review_body[i]), 0)  
    #print(review_body)

    # 2. 이외의 숫자, 특수문자 제거
    review = re.sub('[^a-zA-Z]', ' ', review)
    #print(review_body)

    # 3. 소문자화
    review = review.lower()
    #print(review_body)

    # 4. tokenize
    review  = review.split()
    #print(review_body)

    # 5. stopword, 한글자 제거
    

    final =[]
    for w in review:
        if w not in stopword_list:
            if len(w)>1:
                final.append(w)
            else : pass
        else: pass    
    print(final)


    # 6. lemmatizing
    final_list = []
    for p in final:
        z = lemmatizer.lemmatize(p)
        final_list.append(z)
    #print(final_list)


    finallist.append(final_list)
print(finallist)
```

```python
# stopwords 를 제거한 토큰들
print(len(str(review_body)),end='')
print("개의 단어에서"),
print(len(finallist),end='')
print("개로 줄였습니다")
```
