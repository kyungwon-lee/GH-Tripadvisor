```python
import pandas as pd
import numpy as np
import csv
```

```python
train = pd.read_csv('hoteldiv.csv')
```

```python
train.describe
```

```python
train.review_body = train.review_body.astype(str)
```

```python
train.review_body[1]
```

```python
for i in range(4998):
    train.review_body[i] = train.review_body[i].replace("'", "")
```

```python
len(train.review_body[0])
```

```python
word = []
for i in range(4999):
    temp = []
    temp = train.review_body[i].split(',')
    for j in range(len(temp)):
        word.append(temp[j])
        
print (len(word))
```

```python
for i in range(345696):
    word[i] = word[i].replace("]", "")
    word[i] = word[i].replace("[", "")
```

```python
word[0]
```

```python
import nltk
text = nltk.Text(word, name='NMSC')

# 중복을 제외한 토큰의 개수
print(len(word))         

# 출현 빈도가 높은 상위 토큰 10개
print(text.vocab().most_common(10))
```
