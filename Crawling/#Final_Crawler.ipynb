```python
score_list = []
review_list = []
num = 1

for i in range(254): #여기 페이지 번호!
        site_1 = 'https://www.tripadvisor.com/Hotel_Review-g60763-d93526-Reviews-'
        site_2 = 'The_Lexington_Hotel_Autograph_Collection-New_York_City_New_York.html'      
 
        driver.get(site_1 + 'or' + str(5*i) + '-' + site_2) 
        #req = requests.get(site_1 + 'or' + str(5*i) + '-' + site_2) 
        driver.implicitly_wait(3)
                
        #여기서부터가 Rate crawler
        #html = req.text
        #parse = bs(html, 'html.parser')
        #id_finder = parse.find_all('div', {"class": "hotels-review-list-parts-RatingLine__bubbles--1oCI4"})
        
        driver.find_element_by_class_name("hotels-review-list-parts-RatingLine__bubbles--1oCI4").click()
        r = driver.page_source
        p = bs(r, 'lxml')
        id_finder = p.find_all('div', {"class": "hotels-review-list-parts-RatingLine__bubbles--1oCI4"})

        for item in id_finder:
            score = 0

            if "ui_bubble_rating bubble_10" in str(item):
                score = 1
                score_list.append(score)
            elif "ui_bubble_rating bubble_20" in str(item):
                score = 2
                score_list.append(score)
            elif "ui_bubble_rating bubble_30" in str(item):
                score = 3
                score_list.append(score)
            elif "ui_bubble_rating bubble_40" in str(item):
                score = 4
                score_list.append(score)
            elif "ui_bubble_rating bubble_50" in str(item):
                score = 5
                score_list.append(score)
            else :
                score_list.append(score)

            #print ("%d. %s점" %(num, str(score)))
            num += 1
        driver.find_element_by_class_name("hotels-review-list-parts-ExpandableReview__reviewText--3oMkH").click()

        review = driver.page_source

        parse = bs(review, 'lxml')
        sent = parse.find_all("q", {"class" : "hotels-review-list-parts-ExpandableReview__reviewText--3oMkH"})
    
        for item in sent:
            review_list.append(item)
```

```python
from openpyxl import Workbook
wb = Workbook()

sheet1 = wb.active
file_name = '(4)_Lexington_Hotel.xlsx'
num_of_files = 1270  #이부분은 총 list의 item 갯수

sheet1.cell(row=1, column=2).value = 'review_body'
sheet1.cell(row=1, column=1).value = 'score'

for row_index in range(2, num_of_files+2):
    sheet1.cell(row=row_index, column=2).value = str(review_list[(row_index-2)])
    sheet1.cell(row=row_index, column=1).value = score_list[(row_index-2)]

wb.save(filename=file_name)
```
